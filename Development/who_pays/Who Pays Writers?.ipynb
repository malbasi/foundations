### This scraper needs to work in a few discrete steps
1) Compile a list of publications and their links<br>
2) Loop through that list and scrape pay rate and descriptions<br>
3) Join all information together into one df for analysis<br>

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support.ui import WebDriverWait
import pandas as pd
```

## Step 1: 
compile a list of publication names and their links 

```python
driver = webdriver.Chrome()
driver.get('http://whopayswriters.com/#/results')
```

```python
#find the list of publications in the left side-bar
pub_name = driver.find_element_by_xpath('//*[@id="results"]/div[2]/div[2]/div[1]/ul')
pub_list = pub_name.find_elements_by_tag_name('a')
for list in pub_list[:10]:
    print(list.text)
len(pub_list)
```

```python
#pull the URL from the pub_list 
publication_list=[]
loop_count = 0
for list in pub_list:
    publications = {}
    #pull the second half of the UR
    link = list.get_attribute('ng-href')
    #append the first, static part of the URL with the second half.
    publications['link'] = "http://whopayswriters.com/" + link
    #grab the publication name
    publications['name'] = list.text
    #save name and link (dict) to list of publications
    publication_list.append(publications)
    loop_count += 1
    print(loop_count, 'out of', len(pub_list))
print(publication_list[:2])
```

```python
#check the top of the list to make sure it worked
publication_list[:5]
```

```python
#save complete list as a dataframe
df = pd.DataFrame(publication_list)
df.head()
```

```python
#save dataframe as a CSV
df.to_csv('who_pays.csv', index = False)
```
